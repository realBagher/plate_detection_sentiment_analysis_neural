```python
from ultralytics import YOLO

# Load the base YOLO12n model (pre-trained weights)
model = YOLO("yolo12n.pt")

# Define a comprehensive search space for hyperparameter tuning.
# Adjust these ranges based on your experience and resource constraints.
search_space = {
    "lr0": (1e-5, 1e-1),             # Initial learning rate: from 1e-5 to 1e-1
    "momentum": (0.8, 0.98),           # Momentum range for the optimizer
    "weight_decay": (1e-5, 1e-3),      # Weight decay (regularization)
    "degrees": (0.0, 45.0),            # Maximum rotation degrees (0 means no rotation, 45 maximum)
    "translate": (0.0, 0.5),           # Translation factor (fraction of image size)
    "scale": (0.1, 0.9),               # Scaling factor (zoom in/out)
    "fliplr": (0.0, 1.0),              # Horizontal flip probability
    "hsv_h": (0.0, 0.1),               # Hue adjustment (fraction)
    "hsv_s": (0.0, 1.0),               # Saturation adjustment (fraction)
    "hsv_v": (0.0, 1.0)                # Brightness (value) adjustment (fraction)
}

print("Defined search space:")
for key, value in search_space.items():
    print(f"  {key}: {value}")

# Run hyperparameter tuning using a small subset of your data.
# Make sure "small_dataset.yaml" is a YAML file that points to a limited number of images for faster iterations.
results = model.tune(
    data="dataset_min.yaml",   # Use a YAML that references a small subset for tuning
    epochs=30,                   # Use fewer epochs during tuning to conserve resources
    iterations=300,              # Total iterations for the tuning process
    optimizer="AdamW",           # Use AdamW optimizer (can be changed as desired)
    space=search_space,          # Pass our defined search space
    plots=True,                  # Generate plots to visualize hyperparameter evolution
    save=True,                   # Save the best hyperparameters found
    val=True,                     # Evaluate on the validation set during tuning
    batch=4
)

print("Hyperparameter tuning complete. Best hyperparameters:")
print(results)
```
